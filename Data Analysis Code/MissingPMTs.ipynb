```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# Directory of ELOSS mapping
file_dir = r"/Users/sean/Desktop/Work/GitFRIB/Main Data/Sorted Mapping.xlsx"

# Read in sorted ELOSS mapping
mapdf = pd.read_excel(file_dir, index_col=0)

# Read in *.photon file
file_photon = r"/Users/sean/Desktop/Work/GitFRIB/sean_59Ni_140MeV/r8520.photon"
col_names = list(range(120))
photon_df = pd.read_table(file_photon, sep=' ', names=col_names, index_col=False )

def zero(df, num_missing):
    pmts = []
    c=1
    while c<=num_missing:
        random_num = np.random.randint(119)
        #print(random_num, c)
        if random_num not in pmts:
            pmts.append(random_num)
            c += 1
        else:
            pass
    df.loc[:,pmts] = 0
    return df

def write(file_name, df):
    with open(file_name, "w") as file:
        for row in range(len(df)):
            rowf = np.array(df[row])
            for col in range(120):
                file.write(str(round(rowf[col])))
                file.write(" ")
            file.write("\n")
    file.close()
```

```python
m1 = zero(photon_df.copy(), 1)
write("m1.dat", m1)
m2 = zero(photon_df.copy(), 2)
write("m2.dat", m2)
m3 = zero(photon_df.copy(), 3)
write("m3.dat", m3)
m4 = zero(photon_df.copy(), 4)
write("m4.dat", m4)
m5 = zero(photon_df.copy(), 5)
write("m5.dat", m5)
m6 = zero(photon_df.copy(), 6)
write("m6.dat", m6)
m7 = zero(photon_df.copy(), 7)
write("m7.dat", m7)
m8 = zero(photon_df.copy(), 8)
write("m8.dat", m8)
m9 = zero(photon_df.copy(), 9)
write("m9.dat", m9)
m10 = zero(photon_df.copy(), 10)
write("m10.dat", m10)

```
