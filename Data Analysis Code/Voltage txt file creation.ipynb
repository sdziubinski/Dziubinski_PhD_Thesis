```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os

map_f = os.getcwd() + '\\ELOSS Channel Mapping.xlsx'
cal_f = os.getcwd() + '\\UV PMT Calibration - Version 6.xlsx'
map_df = pd.read_excel(map_f, header = 2, usecols = list(range(0,16,1)), nrows = 128)
cal_df = pd.read_excel(cal_f, header = 0, usecols = list(range(1,17,1)), nrows = 240, sheet_name = [0,1])
```

```python
led1_short = cal_df[0].iloc[::3]
led2_short = cal_df[0].iloc[1::3]
led3_short = cal_df[0].iloc[2::3]
led1_long = cal_df[1].iloc[::3]
led2_long = cal_df[1].iloc[1::3]
led3_long = cal_df[1].iloc[2::3]


led1_short.reset_index(drop = True, inplace = True)
led2_short.reset_index(drop = True, inplace = True)
led3_short.reset_index(drop = True, inplace = True)
led1_long.reset_index(drop = True, inplace = True)
led2_long.reset_index(drop = True, inplace = True)
led3_long.reset_index(drop = True, inplace = True)

led2_short.loc[:,"PMT#"],led2_short.loc[:,"Location"]  = led1_short.loc[:,"PMT#"], led1_short.loc[:,"Location"]
led3_short.loc[:,"PMT#"],led3_short.loc[:,"Location"]  = led1_short.loc[:,"PMT#"], led1_short.loc[:,"Location"]
led2_long.loc[:,"PMT#"],led2_long.loc[:,"Location"]  = led1_long.loc[:,"PMT#"], led1_long.loc[:,"Location"]
led3_long.loc[:,"PMT#"],led3_long.loc[:,"Location"]  = led1_long.loc[:,"PMT#"], led1_long.loc[:,"Location"]

led1 = pd.concat([led1_short, led1_long], ignore_index=True)
led2 = pd.concat([led2_short, led2_long], ignore_index=True)
led3 = pd.concat([led3_short, led3_long], ignore_index=True)
```

```python
cal_voltages = led3.loc[:,['PMT#','Location','Fit Voltage','Adjusted Voltage']]
for i in range(len(cal_voltages['Adjusted Voltage'])):
    if np.isnan(cal_voltages.loc[i,'Adjusted Voltage']) == False:
        cal_voltages.loc[i,'Fit Voltage'] = cal_voltages.loc[i,'Adjusted Voltage']
        
        
cal_voltages = cal_voltages.drop('Adjusted Voltage', axis=1)
```

```python
HVmap_df = map_df.loc[:,['Slot','HV Channel','PMT Serial ']]
HVmap_df = HVmap_df.sort_values(by='PMT Serial ')

cal_voltages_sorted = cal_voltages.sort_values(by='PMT#')
```

```python
HVmap_df.reset_index(drop=True, inplace=True)
cal_voltages_sorted.reset_index(drop=True, inplace=True)
volts = HVmap_df.join(cal_voltages_sorted)
```

```python
volts = volts.set_index('PMT#')
```
